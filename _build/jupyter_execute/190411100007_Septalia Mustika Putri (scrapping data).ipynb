```python
#menampilkan webfix
print (webfix)
```

```python
#mengambil data tertentu dari website
p = webfix.find_all('div', {'class', 'articles--iridescent-list'})
```

```python
#mengambil data tertentu dari website
p.find_all('article', {'class', 'articles--iridescent-list--item articles--iridescent-list--text-item'})
```

```python
#mengambil data tertentu dari website
populer = p.find('div', {'class', 'articles--iridescent-list'})
```

```python
berita = populer.find_all('article', {'class', 'articles--iridescent-list--item articles--iridescent-list--text-item'})
```
