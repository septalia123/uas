```python
import requests
from bs4 import BeautifulSoup
```

```python
web = requests.get('https://www.jawapos.com/berita-hari-ini/')
```

```python
#memnuktikan jika berhasil mengakses web
web
```

```python
#melihat isi website
web.content
```

```python
#agar tampilan content web tertata rapi
webfix = BeautifulSoup( web.content)
```

```python
#menampilkan webfix
print (webfix)
```

```python
#mengambil data tertentu dari website
berita = webfix.find('div', {'class', 'post-list'})
```

```python
print (berita)
```

```python
terkini = berita.find_all('div', {'class', 'post-list__item'})
```

```python
print (terkini)
```

```python
for each in terkini:
    judul = each.find('h3', {'class', 'post-list__title'}).text
    kategori = each.find('a', {'class', 'category text-uppercase'}).text
    isi = each.find('p').text
    waktu = each.find('div', {'class', 'post-list__time'}).text
    
    
    print (kategori)
    print (waktu)
    print (judul)
    print (isi)
    
    print ('='*20)
```
